```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://coinmarketcap.com/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', {'id': 'cryptocurrency'})

rows = table.find_all('tr')[1:11] # Select the first 10 rows (excluding the header)

data = []
for row in rows:
    rank = row.find('td', {'class': 'cmc-table__cell--rank'}).text.strip()
    name = row.find('p', {'class': 'cmc-table__cell--sort-by__name'}).text.strip()
    symbol = row.find('p', {'class': 'cmc-table__cell--sort-by__symbol'}).text.strip()
    price = row.find('div', {'class': 'cmc-table__cell--sort-by__price'}).text.strip()
    market_cap = row.find('p', {'class': 'cmc-table__cell--sort-by__market-cap'}).text.strip()

    data.append([rank, name, symbol, price, market_cap])

df = pd.DataFrame(data, columns=['Rank', 'Name', 'Symbol', 'Price', 'Market Cap'])
print(df)
```
